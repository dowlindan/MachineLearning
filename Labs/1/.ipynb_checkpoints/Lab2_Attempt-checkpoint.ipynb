### Our Data
Create a *random* data matrix with 1600 observations and 20 features.   The range of values are up to you, however these features should be *continuous*.

```python
import numpy as np
X = np.random.default_rng().random((1600,20))
X = X * 100
print(X)
print(X.shape)
```

### Principal Component Analysis
Reduce the demonionality of your data to 2D using PCA (using the two most relevant eigenvectors).  You may use *eig*, *eigh*, or *svd* to do this, but make sure to read the function's documentation for usage.   Finally, visualize your 2D data using matplotlib

```python
#Reduce dimensionality to 2D using PCA
import matplotlib.pyplot as plt
Xz = X - np.mean(X, axis=0)
print(Xz)
Xcov = (Xz.T @ Xz)/(Xz.shape[0]-1) # covariance matrix
[vals, vecs] = np.linalg.eig(Xcov)
W = vecs[:, np.argsort(-vals)[:2]] # projection matrix

Z = Xz @ W # projected data
print(Z)
print(Z.shape)
#TODO Visualize 2D data using matplotlib
plt.plot(Z[:,0], Z[:,1], 'o')
```

### Reconstruction
Finally, let's see how we can reconstruct an observation.   In this section you will use **all** the principle components to project your first observation to its new *N* dimensional space, then *reconstruct* that observation.

```python
##TODO
#Project your first observation into your new space using all the components (note: the dimensionality will not change).
W20 = vecs
X1projected = Xz[0,:] @ W20
#Then reconstruct your observation
X1reconstructed = X1projected @ W20.T
#Print out the original and reconstructed observations
print("Original:", Xz[0,:])
print("Reconstructed:", X1reconstructed)
```
