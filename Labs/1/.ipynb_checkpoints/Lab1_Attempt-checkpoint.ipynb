## Our Data ##
In the next section, I'll define our data as a 2D numpy array:

$X = \begin{bmatrix} -1 & 2 & 3.2\\ 3 & 0 & 8 \end{bmatrix}$

````python
import numpy as np
X = np.array([[-1,2,3.2],[3,0,8]], dtype=float)
print(X)
print(X.shape)
````

## Converting to Binary ##
First let's convert this data to binary.  Use the mean of each column, to convert the features to binary such that if a column's value is below the mean of that column, it get's set to zero, otherwise it gets set to one.

````python
print("X: ")
print(X)

m = np.mean(X, axis=0)
print("Mean of each column: ")
print(m)

bin_arr = (X >= m).astype(int)
print("Binary array: ")
print(bin_arr)
````

## Standardizing Data ##
While there are many ways to standardize your data, let's use z-scoring.

Recall that to z-score our data we need the mean and standard deviation of each column.   As a sanity check, check out the mean and standard deviation of each column before and after standardizing.   **Remember** get the mean and std of each column, not the entire matrix.   Also don't forget to use *ddof=1* for your std.

````python
print("X: ")
print(X)

#Get mean and std of each column and print them out

mean = np.mean(X, axis=0)
print("Mean: ")
print(mean)

stddev = np.std(X, axis=0, ddof=1)
print("Std: ")
print(stddev)

Xz = X - mean
print("X - mean: ")
print(Xz)

Xz = Xz / stddev
print("(X - mean) / stddev [z-scored X data]): ")
print(Xz)
#Z-score data

newMean = np.mean(Xz, axis=0)
print("New Z-Scored Mean: ")
print(newMean)

stddev = np.std(Xz, axis=0, ddof=1)
print("New Z-Scored Std: ")
print(stddev)

#Print new mean and std of each column.
````
